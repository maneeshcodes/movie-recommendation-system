```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
movies = pd.read_csv('../data/tmdb_5000_movies.csv')
credits = pd.read_csv('../data/tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
credits.head()
```

```python
movies = movies.merge(credits, on='title')
```

```python
movies.shape
```

```python
movies.head(2)
```

```python
movies.info()
```

```python
movies = movies[['movie_id','title','overview','keywords','genres','cast','crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
```

```python
def convert(obj):
    lst=[]
    for i in ast.literal_eval(obj):
        lst.append(i['name'])

    return lst
        
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert_cast(obj):
    lst=[]
    counter=0
    for i in ast.literal_eval(obj):
     if counter != 3:
        lst.append(i['name'])
        counter+=1
     else:
        break
    return lst
    
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies.head(2)
```

```python
def fetch_director(obj):
    lst=[]
    for i in ast.literal_eval(obj):
      if i['job'] == 'Director':
        lst.append(i['name'])
        break

    return lst
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head(3)
```

```python
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(2)
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['keywords'] + movies['genres'] + movies['cast'] + movies['crew']
```

```python
movies.head(3)
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df['tags'] = new_df['tags'].str.lower()
```

```python
new_df['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english',ngram_range=(1, 2)) 
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), 
                        reverse=True, 
                        key=lambda x: x[1])[1:6]
    
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
import pickle

pickle.dump(new_df, open('../saved/movies.pkl', 'wb'))
pickle.dump(similarity, open('../saved/similarity.pkl', 'wb'))
```

```python
recommend('Batman Begins')
```
